## Working with Pinecone Indexes.

```python
from pinecone import  Pinecone

pc = Pinecone(api_key=os.getenv('PINECONE_API_KEY'))

pc.list_indexes()
```

```python
pc.describe_index('langchain')
```

```python
pc.list_indexes().names()
```

```python
from pinecone import ServerlessSpec

index_name = 'langchain'

if index_name not in pc.list_indexes().names():
    print(f'Creating index: {index_name}')
    pc.create_index(
        name=index_name,
        dimension=1536,
        metric='cosine',
        spec=ServerlessSpec(
            cloud='aws',
            region='us-east-1',
        )

    )
    print(f'Index {index_name} created successfully.')
else:
    print(f'Index {index_name} already exists.')

```

```python
index_name = 'langchain'

if index_name in pc.list_indexes().names():
    print(f'Deleting index: {index_name}')
    pc.delete_index(name=index_name)
    print(f'Index {index_name} deleted successfully.')
else:
    print(f'Index {index_name} does not exist.')
```

```python
index = pc.Index(index_name)
index.describe_index_stats()
```

## Working with Vectors

```python
import random

vectors = [
    [random.random() for _ in range(1536)] for v in range(5)
]

ids = list('abcde')

index_name = 'langchain-aws'

index = pc.Index(index_name)

index.upsert(vectors=zip(ids, vectors))
```

```python
index.upsert(vectors=[('c',[0.5]*1536)])
```

```python
index.fetch(ids=['c','d'])
```

```python
index.delete(ids=['c','b'])
```

```python
index.describe_index_stats()
```

```python
index.fetch(ids=['c'])
```

```python
index.fetch(ids=['a'])
```

```python
# query

query_vector = [random.random() for _ in range(1536)]
```

```python
index.query(
    vector=query_vector,
    top_k=3,
    include_values=False,
)
```

### Pinecone Namespaces

```python
index = pc.Index('langchain')

import random
vectors = [
    [random.random() for _ in range(1536)] for v in range(5)
]
ids = list('abcde')

index.upsert(vectors=zip(ids, vectors))
```

```python
vectors = [
    [random.random() for _ in range(1536)] for v in range(3)
]
ids = list('xyz')

index.upsert(vectors=zip(ids, vectors), namespace='first-namespace')
```

```python
vectors = [
    [random.random() for _ in range(1536)] for v in range(3)
]

ids = list('xyz')
index.upsert(vectors=zip(ids, vectors), namespace='second-namespace')
```

```python
index.describe_index_stats()
```

```python
index.fetch(ids=['x'])
```

```python
index.fetch(ids=['x'], namespace='first-namespace')
```

```python
index.delete(ids=['x'], namespace='first-namespace')
```

```python
index.delete(delete_all=True, namespace='first-namespace')
```
